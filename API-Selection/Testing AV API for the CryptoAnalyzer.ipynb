# Testing the Alpha Vantage API for CryptoCurrency Daily Data

## Libraries

```python
import requests
import pandas as pd
```

##Example output:
https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=CNY&apikey=demo

```python
# Replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=CNY&apikey=demo'
r = requests.get(url)
data = r.json()
print(data)
```

```python
meta_data = pd.DataFrame.from_dict(data['Meta Data'], orient="index")
meta_data
```

```python
daily = pd.DataFrame.from_dict(data['Time Series (Digital Currency Daily)'], orient="index")
daily
```
